```python
import matplotlib.pyplot as plt
```

```python
# Checking age for outliers
phd.boxplot(column=['RestBP'])
plt.show()
```

```python
# checking RestBP median
print(phd["RestBP"].median())
```

```python
# Calculating quantiles
# Code is Courtesy of Gemini
Q1 = phd["RestBP"].quantile(0.25)
Q3 = phd["RestBP"].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound (Outliers below): {lower_bound}")
print(f"Upper Bound (Outliers above): {upper_bound}")
outliers = phd[
    (phd["RestBP"] < lower_bound) | (phd["RestBP"] > upper_bound)
]

print("\n--- Outliers Identified ---")
if outliers.empty:
    print(f"No outliers found in the '{"RestBP"}' column based on the IQR method.")
else:
    print(outliers)
```

```python
# Replacing Outliers with median
# Code courtesy of gemini
phd.loc[phd["RestBP"] > upper_bound, "RestBP"] = phd["RestBP"].median()
phd.loc[phd["RestBP"] < lower_bound, "RestBP"] = phd["RestBP"].median()
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_confusion_matrix(y_true, y_pred, model_name, class_names):
    """Generates and plots a confusion matrix as a heatmap."""
    
    # 1. Calculate the Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # 2. Convert to DataFrame for easier visualization with class labels
    cm_df = pd.DataFrame(cm, 
                         index = class_names, 
                         columns = class_names)

    # 3. Create the Heatmap Plot
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm_df, 
                annot=True,        # Show the numbers in the cells
                fmt='d',           # Format as integers
                cmap='Blues',      # Use a blue color scheme
                cbar=False)        # Don't show the color bar
    
    plt.title(f'Confusion Matrix: {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Assuming your target variable classes are decoded:
# Replace these with your actual class names (e.g., ['No Disease', 'Disease'])
class_names = ['No disease', 'Disease']
```

```python
plot_confusion_matrix(y_test, y_pred_logreg, 'Logistic Regression', class_names)
```

```python
plot_confusion_matrix(y_test, y_pred_dt, 'Decision Tree', class_names)
```

```python
plot_confusion_matrix(y_test, y_pred_svm, 'SVM', class_names)
```

## Hyperparameter Tuning

```python
# I'll use randomized search first to locate the rough area of the optimal hyperparameters
# Then I'll follow up with a grid search to find them in the narrower area.
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import recall_score, make_scorer
from scipy.stats import randint as sp_randint
```

```python
# Defines the set of parameters for randomized search to use
param_dist = {
    'n_estimators': sp_randint(50, 300),          # Random integers between 50 and 300
    'max_depth': [10, 20, 30, None],             # Specific discrete values
    'min_samples_split': sp_randint(2, 11),      # Random integers between 2 and 10
    'criterion': ['gini', 'entropy']
}
```

```python
# Defines the number of random experiments for randomized search to run
N_ITER = 40
```

```python
scorer = make_scorer(recall_score, average='weighted')
```

```python
N_ITER_RANDOM = 50 # Number of random combinations to test

random_search = RandomizedSearchCV(
    estimator=rf_model, 
    param_distributions=param_dist, 
    n_iter=N_ITER_RANDOM, 
    scoring=scorer, 
    cv=5, 
    verbose=2, 
    random_state=42, 
    n_jobs=-1
)
```

```python
print("--- Starting Randomized Search (Broad Exploration) ---")
random_search.fit(x_train, y_train)

best_random_params = random_search.best_params_
print("\nRandomized Search Best Params:", best_random_params)
```

```python
# Now running grid search based on results from randomized search
```

```python
# Assuming best_random_params is available after the Randomized Search

best_min_split = best_random_params['min_samples_split']

# 1. Calculate the lower bound, ensuring it's at least 2
lower_bound = max(2, best_min_split - 1)

# 2. Calculate the upper bound
upper_bound = best_min_split + 1

# 3. Create the list, use a set to ensure uniqueness, and sort it
split_list_final = sorted(list(set([lower_bound, best_min_split, upper_bound])))


# --- Safety Check for max_depth (to prevent the previous NoneType error) ---
best_depth = best_random_params['max_depth']
depth_list_final = []

if best_depth is None:
    depth_list_final = [None, 30, 40]
else:
    # Bracket the best depth, ensuring values are > 0 and include None for safety
    depth_list_final = [best_depth - 5, best_depth, best_depth + 5]
    depth_list_final = [d for d in depth_list_final if d is not None and d > 0]
    depth_list_final.append(None)
    depth_list_final = sorted(list(set(depth_list_final)))


# --- Define the FOCUSED Grid Dictionary ---
focused_param_grid = {
    # Bracket the best n_estimators
    'n_estimators': list(range(best_random_params['n_estimators'] - 25, 
                               best_random_params['n_estimators'] + 50, 
                               25)),
                               
    # Use the cleaned list
    'max_depth': depth_list_final,
    
    # Use the cleaned list
    'min_samples_split': split_list_final,
    
    'criterion': [best_random_params['criterion']] 
}

# Print the resulting grid to verify the values before running GridSearch
print("\nGenerated Focused Grid:")
print(focused_param_grid)

# You can now safely run GridSearch with this dictionary
# grid_search = GridSearchCV( ... param_grid=focused_param_grid, ... )
```

```python
grid_search = GridSearchCV(
    estimator=rf_model, 
    param_grid=focused_param_grid, 
    scoring=scorer, 
    cv=5, 
    verbose=2, 
    n_jobs=-1 # Use all processor cores
)

# 2. Run the search on the training data
print("\n--- Starting Focused Grid Search (Fine Tuning) ---")
grid_search.fit(x_train, y_train)

# 3. Output the final best parameters found
print("\nGrid Search Final Best Parameters:", grid_search.best_params_)
print("Grid Search Final Best Score (Recall):", grid_search.best_score_)

# 4. Access the best trained model directly
best_rf_model = grid_search.best_estimator_
```

```python
## Creating a new RF model with the optimized parameters
best_rf_model = grid_search.best_estimator_
print("Final Optimal Parameters Used:")
print(best_rf_model.get_params())
```
