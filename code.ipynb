```python
# Locating rows whose Thal columns have null in them 
phd["Thal"].isna()
```

```python
# Using null masking based on gemini advice
null_mask = phd['Thal'].isnull()
rows_with_nulls = phd[null_mask]
print("\n--- Rows with Null Data (The Result) ---")
print(rows_with_nulls)
```

```python
# Confirming indices are correct
# found code at https://www.geeksforgeeks.org/pandas/select-rows-columns-by-name-or-index-in-pandas-dataframe-using-loc-iloc/
t1 = phd.loc[:87]
t2 = phd.loc[:266]
```

```python
# Deleting the rows using drop()
# Learned drop here: https://docs.vultr.com/python/third-party/pandas/drop
phd = phd.drop([87, 266])
```
