```python

# checking RestBP median
print(phd["RestBP"].median())
```

```python
# Calculating quantiles
# Code is Courtesy of Gemini
Q1 = phd["RestBP"].quantile(0.25)
Q3 = phd["RestBP"].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound (Outliers below): {lower_bound}")
print(f"Upper Bound (Outliers above): {upper_bound}")
outliers = phd[
    (phd["RestBP"] < lower_bound) | (phd["RestBP"] > upper_bound)
]

print("\n--- Outliers Identified ---")
if outliers.empty:
    print(f"No outliers found in the '{"RestBP"}' column based on the IQR method.")
else:
    print(outliers)
```

```python
# Replacing Outliers with median
# Code courtesy of gemini
phd.loc[phd["RestBP"] > upper_bound, "RestBP"] = phd["RestBP"].median()
phd.loc[phd["RestBP"] < lower_bound, "RestBP"] = phd["RestBP"].median()
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_confusion_matrix(y_true, y_pred, model_name, class_names):
    """Generates and plots a confusion matrix as a heatmap."""
    
    # 1. Calculate the Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # 2. Convert to DataFrame for easier visualization with class labels
    cm_df = pd.DataFrame(cm, 
                         index = class_names, 
                         columns = class_names)

    # 3. Create the Heatmap Plot
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm_df, 
                annot=True,        # Show the numbers in the cells
                fmt='d',           # Format as integers
                cmap='Blues',      # Use a blue color scheme
                cbar=False)        # Don't show the color bar
    
    plt.title(f'Confusion Matrix: {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Assuming your target variable classes are decoded:
# Replace these with your actual class names (e.g., ['No Disease', 'Disease'])
class_names = ['No Disease', 'Disease']
```
