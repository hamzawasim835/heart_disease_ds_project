## Locating and Handling outliers

```python
import matplotlib.pyplot as plt
```

```python
# Checking age for outliers
phd.boxplot(column=['Age'])
plt.show()
```

```python
# Checking age for outliers
phd.boxplot(column=['RestBP'])
plt.show()
```

```python
# checking RestBP median
print(phd["RestBP"].median())
```

```python
# Calculating quantiles
# Code is Courtesy of Gemini
Q1 = phd["RestBP"].quantile(0.25)
Q3 = phd["RestBP"].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound (Outliers below): {lower_bound}")
print(f"Upper Bound (Outliers above): {upper_bound}")
outliers = phd[
    (phd["RestBP"] < lower_bound) | (phd["RestBP"] > upper_bound)
]

print("\n--- Outliers Identified ---")
if outliers.empty:
    print(f"No outliers found in the '{"RestBP"}' column based on the IQR method.")
else:
    print(outliers)
```

```python
# Replacing Outliers with median
# Code courtesy of gemini
phd.loc[phd["RestBP"] > upper_bound, "RestBP"] = phd["RestBP"].median()
phd.loc[phd["RestBP"] < lower_bound, "RestBP"] = phd["RestBP"].median()
```

```python
# Visually confirming result
phd.boxplot(column=['RestBP'])
plt.show()
```

```python
#  Handling Cholestrol
# 200 > x is good, 200 to 239 is borderline to moderately elevated, 240 is high
# Since there is only one cholestrol listed, it's safe to assume data is about total cholestrol, not one type
# of it
# Running a visual check
phd.boxplot(column=['Chol'])
plt.show()
```

```python
disease_by_gender = phd.groupby('Sex')['Target'].sum()

print("Number of individuals with disease (target = 1) by gender:")
print(disease_by_gender)
```

```python
# Vastly more men than women
# counts total number of sick people. Courtesy Gemini
total_sick = phd['Target'].sum()

# Number of sick men: Filter for Sex = 1 AND target = 1
sick_men = phd[(phd['Sex'] == 1) & (phd['Target'] == 1)].shape[0]

# Calculate the percentage
if total_sick > 0:
    percentage_sick_men = (sick_men / total_sick) * 100
    
    print(f"Total number of sick people: {total_sick}")
    print(f"Number of sick men: {sick_men}")
    print(f"\nPercentage of sick men to the whole amount of sick people: {percentage_sick_men:.2f}%")
```

```python
# Observation: 81.88 % of sick people are men.
```

```python
# Checking average age of sick men versus women
# Code Courtesy Gemini
# 1. Filter the DataFrame to include only the sick individuals (target = 1)
sick_df = phd[phd['Target'] == 1]

# 2. Calculate the OVERALL average age of all sick people
overall_avg_age = sick_df['Age'].median()

# 3. Calculate the average age of sick men and sick women
# This groups the sick individuals (sick_df) by 'Sex' and finds the mean 'Age' for each group.
avg_age_by_sex = sick_df.groupby('Sex')['Age'].median()

print(f"Overall Average Age of all sick individuals: {overall_avg_age:.2f} years")
print("\nAverage Age of sick individuals, broken down by Sex:")
print(avg_age_by_sex.to_string())
```

```python
# Checking the relationship between blood pressure and sex
import seaborn as sns

# --- Set the correct column names ---
BLOOD_PRESSURE_COLUMN = 'RestBP' # <--- REPLACE WITH YOUR ACTUAL BLOOD PRESSURE COLUMN NAME
GENDER_COLUMN = 'Sex'

# Set the style
sns.set_style("whitegrid")

# Create the boxplot
plt.figure(figsize=(8, 6))

# Generate the boxplot: 'x' is the categorical variable (Sex), 'y' is the numerical variable (Blood Pressure)
sns.boxplot(
    x=phd[GENDER_COLUMN], 
    y=phd[BLOOD_PRESSURE_COLUMN], 
    data=phd,
    palette={'0': 'skyblue', '1': 'salmon'} # Adjust colors if necessary
)

# Label the axes and title (assuming 0=Female, 1=Male)
# Adjust these labels to match the actual representation in your 'Sex' column
plt.title('Resting Blood Pressure Distribution by Sex')
plt.xlabel(f'{GENDER_COLUMN} (0: Female, 1: Male)') 
plt.ylabel(BLOOD_PRESSURE_COLUMN)

# Save the plot
plt.savefig('blood_pressure_boxplot_by_sex.png')
plt.show()
plt.close()
print("Boxplot saved as blood_pressure_boxplot_by_sex.png")
```
